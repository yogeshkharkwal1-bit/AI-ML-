## Linear Regression using Gredian Descent

```python
# Linear Regresion
class LinearRegression:
    # we define every theta into object perameeters self
    def __init__ (self,learning_rate=0.01,n_iter=1000):
        # perameters that we need
        self.bias= None
        self.weights= None
        self.lr=learning_rate
        self.n_iter=n_iter

    def fit(self,X ,y):
        m,n= X.shape  #sample of data ,numbers of features
        # STEP 1- initilize perameeters that we need
        self.bias=0
        self.weights=np.zeros(n) # return an array of zeros

        # GREDIENT DESCENT
        for i in range (self.n_iter):  
             
            # STEP2-  calclate y predict
            y_pred=self.bias+ np.dot(X,self.weights)

            # STEP3 - calcualte gredients
            db=(1/m)*np.sum(y_pred- y)
            dw=(1/m)*np.dot(X.T,(y_pred- y))

            # STEP4- apply conversion theorum for perameters update
             
            # self.bias=self.bias - (self.lr*db)
            # self.weight=self.weight - (self.lr*dw)
         
            self.bias -= self.lr*db
            self.weights -= self.lr*dw

         
    def predict(self,X):   
         y_pred=self.bias + np.dot(X,self.weights)
         return y_pred
```

```python
X=np.array([[1],[2],[3],[4],[5]])
y=np.array([2,4,6,8,10])

model=LinearRegression()
model.fit(X,y)

y_pred=model.predict(X)
print(y_pred)

```

## Ordinary least squares

```python
class LinearRegressionOLS:
    # we define every theta into object perameeters self
    def __init__ (self,n_iter=1000):
        # perameters that we need
        self.bias= None
        self.weights= None

    def fit(self,X,y):
        m,n=X.shape
        
        X_updated=np.c_[np.ones((m, 1)),X]

        # normal equations
        theta=np.dot(np.linalg.inv(np.dot(X_updated.T,X_updated)),np.dot(X_updated.T,y))
        # theta=np.dot(np.linalg.inv(X_updated.T @ X_updated) @ (X_update.T @ y))

        self.bias=theta[0]
        self.weights=theta[1:]
    def predict(self,X):  
         y_pred=self.bias + np.dot(X,self.weights)
         return y_pred
```

```python
X=np.array([[1],[2],[3],[4],[5]])
y=np.array([2,4,6,8,10])

model=LinearRegressionOLS()
model.fit(X,y)

y_pred=model.predict(X)
print(y_pred)
```
