Logistic Regression

```python
# Logistic Regresion
class LogisticRegression:
    # we define every theta into object perameeters self
    def __init__ (self,learning_rate=0.1,n_iter=1000):
        # perameters that we need
        self.bias= None
        self.weights= None
        self.lr=learning_rate
        self.n_iter=n_iter

    def _sigmoid(self,z):
        return(1/(1+np.exp(-z)))
        
    def fit(self,X ,y):
        m,n= X.shape  #sample of data ,numbers of features
        
        # STEP 1- initilize perameeters that we need
        self.bias=0
        self.weights=np.zeros(n) 

        # GREDIENT DESCENT
        for i in range (self.n_iter):  
             
            # STEP2-  calclate y predict
            z=self.bias+ np.dot(X,self.weights)
            y_pred=self._sigmoid(z)
            
            # STEP3 - calcualte gredients
            db=(1/m)*(y_pred - y)
            dw=(1/m)*np.dot(X.T,(y_pred- y))

            # STEP4- apply conversion theorum for perameters update
            self.bias -= self.lr*db
            self.weights -= self.lr*dw

    def get_probabilities(self,X):
            z=self.bias+ np.dot(X,self.weights)
            return self._sigmoid(z)
        
    def predict(self,X,threshold=0.5):   
        probability=self.get_probabilities(X)
        y_pred=probability >= threshold
        return y_pred.astype(int)
        
```
