## SVR== SUPORT VECTOR REGRESOR

```python
from sklearn.svm import SVR
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score,accuracy_score
```

```python
df=datasets.load_diabetes(as_frame=True).frame
```

```python
df.head()
```

```python
X=df.drop("target",axis=1)
y=df["target"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(
    X,y,test_size=0.3,random_state=42
)
```

## change the y dimention

```python
y_scaler = StandardScaler()

y_train_scaled=y_scaler.fit_transform(y_train.values.reshape(-1,1)).ravel()
y_test_scaled=y_scaler.transform(y_test.values.reshape(-1,1)).ravel()
```

```python
model=SVR()
model.fit(X_train,y_train_scaled)
```

```python
y_test_pred_scaled=model.predict(X_test)
y_train_pred_scaled=model.predict(X_train)
```

```python
print("r2_test_score :",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_train_score :",r2_score(y_train_scaled,y_train_pred_scaled))
```

```python
model=SVR(kernel="linear")
model.fit(X_train,y_train_scaled)

y_test_pred_scaled=model.predict(X_test)
y_train_pred_scaled=model.predict(X_train)

print("r2_test_score :",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_train_score :",r2_score(y_train_scaled,y_train_pred_scaled))
```

```python
model=SVR(kernel="poly")
model.fit(X_train,y_train_scaled)

y_test_pred_scaled=model.predict(X_test)
y_train_pred_scaled=model.predict(X_train)

print("r2_test_score :",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_train_score :",r2_score(y_train_scaled,y_train_pred_scaled))
```

```python
model=SVR(kernel="sigmoid")
model.fit(X_train,y_train_scaled)

y_test_pred_scaled=model.predict(X_test)
y_train_pred_scaled=model.predict(X_train)

print("r2_test_score :",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_train_score :",r2_score(y_train_scaled,y_train_pred_scaled))
```

## use gridsearchCV for best hyper perameters to solve overfitting

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid= {
    "C":[1,2,5,10,15,20],
    "kernel":["linear","poly","rbf"],
    "epsilon":[0.01,0.1,0.2,0.3,0.5]
}
```

```python
svr=SVR()
grid_search = GridSearchCV(svr,param_grid,scoring="r2",cv=5)
grid_search.fit(X_train,y_train_scaled)

```

```python
print(grid_search.best_params_)
```

```python
best_model=SVR(C=15,kernel="linear",epsilon=0.1)
best_model.fit(X_train,y_train_scaled)

y_test_pred_scaled=best_model.predict(X_test)
y_train_pred_scaled=best_model.predict(X_train)

print("r2_test_score :",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_train_score :",r2_score(y_train_scaled,y_train_pred_scaled))
```

## LINEAR SVR

```python
from sklearn.svm import LinearSVR

best_model=LinearSVR(C=15,epsilon=0.1,max_iter=10000)
best_model.fit(X_train,y_train_scaled)

y_test_pred_scaled=best_model.predict(X_test)
y_train_pred_scaled=best_model.predict(X_train)

print("r2_test_score :",r2_score(y_test_scaled,y_test_pred_scaled))
print("r2_train_score :",r2_score(y_train_scaled,y_train_pred_scaled))
```
